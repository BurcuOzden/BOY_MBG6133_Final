```python
#Â Import required libraries
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Question 1

```python
# Random numpy array is generated at 100x100 size and the numbers are between 0 and 100.
randomarray=np.random.randint(101, size=(100, 100))
# Print the array size to control it.
print(randomarray.shape)
# Print the generated array to control it.
print (randomarray)
```

```python
# Heatmap of generated random array. It is saved with the name of "heatmap_randomarray" in pdf format at 300 dpi resolution.
plt.figure(figsize=(14,10))
ax=sns.heatmap(randomarray, vmin=0, vmax=100, cmap="PiYG")
plt.savefig("heatmap_randomarray.pdf", dpi=300, format="pdf")
```

# Question 2

```python
# Even and odd numbers in the random array are categorized as true and false. 
# If the number is odd it is labeled as true and otherwise it is labeled as false.
even_odd_array=randomarray%2==1
print(even_odd_array)

# False/True array representation is converted to 0 and 1 matrix. 
# False and True cells are represented with 0 and 1, respectively.
even_odd_array_int=even_odd_array.astype(int)
print(even_odd_array_int)
```

```python
# Heatmap of generated 0-1 array. It is saved with the name of "heatmap_even_odd" in pdf format at 300 dpi resolution.
plt.figure(figsize=(14,10))
ax=sns.heatmap(even_odd_array_int, vmin=0, vmax=1, cmap="tab20")
plt.savefig("heatmap_even_odd.pdf", dpi=300, format="pdf")
```
